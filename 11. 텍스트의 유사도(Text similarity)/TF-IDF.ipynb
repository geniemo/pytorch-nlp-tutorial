```python
from math import log

import pandas as pd
```

```python
docs = [
    "먹고 싶은 사과",
    "먹고 싶은 바나나",
    "길고 노란 바나나 바나나",
    "저는 과일이 좋아요",
]
```

```python
vocab = list(set(w for doc in docs for w in doc.split()))
vocab.sort()
vocab
```

```python
N = len(docs)


def tf(t, d):
    return d.count(t)


def idf(t):
    df = 0
    for doc in docs:
        df += t in doc
    return log(N / (df + 1))


def tfidf(t, d):
    return tf(t, d) * idf(t)
```

```python
result = []

for i in range(N):
    result.append([])
    d = docs[i]
    for j in range(len(vocab)):
        t = vocab[j]
        result[-1].append(tf(t, d))

tf_df = pd.DataFrame(result, columns=vocab)
tf_df
```

```python
result = []
for j in range(len(vocab)):
    t = vocab[j]
    result.append(idf(t))

idf_df = pd.DataFrame(result, index=vocab, columns=["IDF"])
idf_df
```

```python
result = []
for i in range(N):
    result.append([])
    d = docs[i]
    for j in range(len(vocab)):
        t = vocab[j]
        result[-1].append(tfidf(t, d))

tfidf_df = pd.DataFrame(result, columns=vocab)
tfidf_df
```
